### 2) 평탄화
- 다차원 배열을 낮은차원으로 변환

#### ㄱ) ravel
- 다차원 배열을 낮은 차원으로 변환
- ravel은 원본의 뷰를 반환하기 때문에 ravel로 반환된 데이터
   를 수정하면 원본 데이터도 수정 된다.

```python
b = np.array([ [[1, 2], [3, 4]], [[1, 2], [3, 4]] ])
print(b)
```

```python
c = b.ravel()
print(c)
print(type(c))
```

image.png

```python
c[3] = 0
print(c)
```

```python
print(b)
```

#### ㄴ) flatten
- 다차원 배열을 낮은 차원으로 변환
- flatten 원본의 복사본을 반환한다
- flatten은 flatten으로 반환된 데이터를 수정하여도 원본데이터는 변하지 않는다.

```python
b = np.array([[1, 2], [3, 4]])
b
```

```python
c = b.flatten()
c
```

```python
c[3] = 0
c
```

```python
b
```

## (2) NumPy 색인

### 1) 색인 (indexing)
- 배열의 각 요소는 axis 인덱스 배열로 참조 할 수 있다.
- 1차원 배열은 1개의 인덱스, 2차원 배열은 2개의 인덱스,
     n차원 배열은 n개의 인덱스로 요소를 참조 할 수 있다.


```python
a = np.arange(1, 5)
a
```

```python
print("a의 0번째 요소 = ", a[0])  
print("a의 1번째 요소 = ", a[1])  
```

```python
b = a.reshape(2,2)
b
```

```python
print(b[0, 0])  
print(b[1][1]) 
```

- 할당

```python
b[0, 1] = 5   
print(b)
```

```python
b[1][0] = 9   
print(b)
```

### 3) boolean indexing
- 배열 각 요소에 조건에 대한 값을 True, False로 지정한 후,
   해당 index의 True값만 추출하는 방법

```python
a = np.arange(1, 21).reshape(4, 5) #2차원 배열
print(a)
```

```python
bool_a = a%2==1  # 몫의 나머지 = 0 일경우 true
print(bool_a)
```

```python
a[bool_a]
```

```python
b = np.logical_and(a > 5 , a < 9)
b
```

```python
a[b]
```

### 4) fancy indexing
- 정수 배열을 사용한 색인을 설명하기 위해 NumPy에서 차용한 
   단어이다.
- ndarray를 index value로 사용하여 값을 추출한다.

**1차원**

```python
array_a = np.arange(1,7)
print(array_a)
array_b = np.array( [3, 1, 2, 0, 1, 5], int )  # 반드시 integer로 선언
print(array_b)
array_a[array_b]  # bracket index, b 배열의 값을 index로 하여 a의 값 추출
```

```python
array_a.take(array_b)
```

**2차원**

```python
array_a = np.arange(1,13).reshape(4,3)
print(array_a)
array_b = np.array( [3, 1, 2, 0, 1], int ) 
array_c = np.array( [1, 2, 1, 0, 2], int ) 
array_a[array_b,array_c]
```

### 5) argmax, argmin
- array내의 최대값 또는 최소값의 인덱스 반환

```python
a = np.array( [4, 1, 0, 4, 11, 10, 17 ] )
print("최대값 index =",np.argmax(a), "최소값 index=",np.argmin(a))
```

**axis 기반의 반환**

```python
a2 = np.random.choice(np.arange(2, 22), 12).reshape(3, 4)
a2
```

```python
np.argmax(a2, axis = 0), np.argmin(a2, axis = 1)
```

## 3. NumPy i/o

### loadtxt, savetxt
 - 텍스트 파일을 불러오고 저장하는 방법

#### save

```python
a = np.arange(1, 80, 2).reshape(8, 5)
a
```

```python
np.savetxt( "my_int.csv" , a, delimiter = "," )
```

```python
np.savetxt( "my_int_2.csv" , a, fmt = '%d', delimiter = "," )
```

#### load

```python
b = np.loadtxt( "./my_int_2.csv", delimiter = ","  )
b[:5]  # python 기본타입 float 형태로 보임
```

```python
b_int = b.astype(int)
b_int[:3]
```

#### npy
- Numpy object (pickle) 형태로 데이터를 저장하고 불러옴
- Binary 형태로 파일 저장

```python
np.save( "npy_test" , arr = b_int )
```

```python
npy_array = np.load( file = "npy_test.npy")
npy_array[:3]
```
