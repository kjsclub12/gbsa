### 비교 연산 – np.where
- if 절과 같은 효과
-         where(조건, True일 때 원하는value, False일 때 원하는 value)

````
a = np.array( [2, 3, 1], float )
np.where( a > 1 , 0, 3 )
````

````
a = np.arange( 3, 10)
print(a)
print(np.where( a > 6 ))  # True 값의 index 값 반환
````

### 비교 연산 – np.isnan, np.isfinite
- np.isnan : null일 경우 True
- np.isfinite : 한정된 수인 경우 True

````
a = np.array( [ 2, np.NaN, np.Inf ], float )
print(a)
print(np.isnan( a ))  # null인 경우 True
````

````
np.isfinite( a )  # 한정된 수인 경우 True
````

## 3.NumPy 성능비교

- jupyter notebook 환경에서 코드의 성능을 볼 수 있는 %timeit 사용
- 구구단 100단연산 비교

### numpy 활용

````
array_a = np.arange(1,101).reshape(-1,1)
array_a
````

````
array_b = np.arange(1,101)
array_b
````

````
array_a * array_b
````

````
%timeit array_a * array_b
````

### list 자료형, for 반목문 사용

````
list_a = range(1, 101) 
list_b = range(1, 101)
````

````
[i * j for i in list_a for j in list_b]
````

````
%timeit [i * j for i in list_a for j in list_b]
````
