- 나누기 나머지 반환 함수

````
def mod(a,b): # a를 b로 나누어 나머지를 반환한다.
    div = int(a/b)
    mod = a - div*b
    return mod

print(mod(3,2))
print(mod(4,2))
print(mod(5,2))
print(mod(6,2))
````

### 정수를 받아 짝수여부를 판별하는 함수 작성

````
def mod_fun(a,b):
    div = int(a/b)
    mod = a - (div*b)
    return mod
````

````
mod_fun(5,3)
````

````
def is_even(n):
    if mod_fun(n,2) == 0:
        return True
    else:
        return False
````
